```python
import pandas as pd
import sodapy as sodapy
from sodapy import Socrata
```

```python
all_results = []
offset=0
while True:
    client = Socrata("data.cityofnewyork.us", None)
    results = client.get("w9ak-ipjd", where="filing_date>'2023-12-31T00:00:00.000'", limit=50000,offset=offset)
    offset = offset + 50000
    print(len(results),offset)
    all_results.extend(results)
    if len(results)<50000:
        break
all_results_df = pd.DataFrame.from_dict(all_results)
all_results_df
```

```python
all_results_df.info()
```
