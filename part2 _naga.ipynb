# Construction Job Applications Data:

```python
import pandas as pd
from sodapy import Socrata

client = Socrata("data.cityofnewyork.us", None)
df_list = []
offset = 0

while len(df_list)<10000:
    results = client.get("ic3t-wcy2", 
                         limit=2000,
                         offset=offset)
    offset += 2000
    df_list.extend(results)

df = pd.DataFrame.from_records(df_list)
df[['job_type', 'gis_census_tract', 'gis_latitude', 'gis_longitude']]
```

```python
import plotnine as p9

plot = (
    p9.ggplot(df)
    + p9.aes(x='job_type')
    + p9.geom_bar()
    + p9.labs(title='Job Type Distribution')
)

plot
```

```python
import pandas as pd
from sodapy import Socrata

client = Socrata("data.cityofnewyork.us", None)
df_list = []
offset = 0

while len(df_list)<10000:
    results = client.get("w2pb-icbu", 
                         limit=2000,
                         offset=offset)
    offset += 2000
    df_list.extend(results)

df_land = pd.DataFrame.from_records(df_list)

df_land[['sale_price', 'borough', 'latitude', 'longitude']]
```
